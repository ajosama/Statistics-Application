# SciPy for Statistics: Problem Statement
1. Generate 1100 numbers, between 90 and 200, using Numpy's arrange function.
2. Perform the usual mean, sd, norm, and fit Bell Curve using Scipy libraries.

### Problem Statement:
1. Let's assume that the data we generate represents the height of the population of age group 16-24 in Canada.
2. Let's fit a curve to it and perform statistical analysis on it.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import norm
from scipy.optimize import curve_fit

import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
steps = (200-90)/1100
print(steps)
```

```python
ar = np.arange(90,200,0.1)
print(ar)
```

```python
plt.figure(figsize = (5,3), dpi=100)
plt.title('Line plot of the array')
plt.plot(ar)
```

```python
print(len(ar),
np.mean(ar),
np.std(ar))
```

```python
data = norm.rvs(ar)   # Random Variance 
```

```python
axis = np.linspace(1,1100,1100)
```

```python
plt.figure(figsize = (7,5), dpi=100)
plt.title('Scatter Plot')
sns.set_style('darkgrid')
sns.scatterplot(data)
```

```python
# fitting function
def func(x,m,c):
    return m*x + c
```

```python
popt, pcov = curve_fit(func,axis,data,p0=[0,40])
```

```python
popt
```

```python
pcov
```

```python
y_pred = func(axis,popt[0],popt[1])
```

```python
plt.figure(figsize = (7,5), dpi=100)
plt.title('Checking how good our line fits.')
plt.scatter(data,axis)
plt.plot(y_pred,axis,color='r')
```

# Advanced Statistics

### Problem Statement:
1. A stock price always fluctuates, but is there a relation between a closing price and its moving averages?
2. An analyst wants to showcase how the stock would move, he has 5 years of stock price data with him on which he needs to perform the analysis.

```python
# Import Libraries
import yfinance as yahooFinance
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
google = yahooFinance.Ticker('GOOG')
```

```python
# 5 years of data
df = google.history(period='5Y')
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
df.describe()
```

```python
# Overview of closing price before calculating DMA's
sns.set_style('darkgrid')
plt.figure(figsize=(7,5),dpi=100)
plt.title('Closing Price')
plt.plot(df['Close'])
```

```python
# Calculate 50DMA and 200DMA
df['FiftyDMA'] = df['Close'].rolling(50).mean()
df['TwoHundredDMA'] = df['Close'].rolling(200).mean()
```

```python
df.describe()
```

```python
df.drop(columns = ['Dividends','Stock Splits'],inplace = True)
```

```python
df.head()
```

```python
# Plot Closing price vs 50DMA vs 200DMA
sns.set_style('darkgrid')
plt.figure(figsize=(7,5),dpi=100)
plt.title('Closing Price vs 50DMA vs 200DMA')
plt.plot(df['Close'],label = 'Close')
plt.plot(df['FiftyDMA'],label = '50DMA')
plt.plot(df['TwoHundredDMA'],label = '200DMA')
plt.legend()
```

## Inferences from the above graph:
1. 50DMA and 200DMA are fairly good indicators of how the stock is moving.
2. If the closing price is above the 50DMA or 200DMA, it is usually means that it is on the uptrend.
3. If the closing price is below the 50DMA or 200DMA, it is usually means that it is in a down trend.
4. Every point where 50DMA and 200DMA intersect, the market usually reverse its trend (High -> Low | Low -> High).
5. The huge dip at the March 2020, indicates the Covid 19 market crash.

# Finding the possible futuristic movement of the stock

```python
# Analyse Correlation between each variables
plt.figure(figsize=(7,7),dpi=100)
sns.heatmap(df.corr(),annot = True)
```
